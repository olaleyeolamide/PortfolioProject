```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#to adjust the configuation of the plots to be created
%matplotlib inline
matplotlib.rcParams ['figure.figsize'] = (12,8)

#read in the data
df = pd.read_csv(r'C:\Users\DELL\Downloads\movies.csv\movies.csv')
```

```python
df
```

```python
#to check for the percentage of missing data in each columns

for col in df.columns:
    missing_data = np.mean(df[col].isnull())
    print(col, (missing_data*100),'%')
```

```python
#to replace missing vales present in 'score', 'budget' and 'gross' column with the mean of the available values

mean = df["score"].mean()
print(mean)
df["score"].replace(np.nan,mean, inplace=True)

mean = df["budget"].mean()
print(mean)
df["budget"].replace(np.nan,mean, inplace=True)

mean = df["gross"].mean()
print(mean)
df["gross"].replace(np.nan,mean, inplace=True)

#to confirm all missing data in the specified columns are now replaced with their mean
for col in df.columns:
    missing_data = np.mean(df[col].isnull())
    print(col, (missing_data*100),'%')
```

```python
#to drop every other missing values

df.dropna()
df.head()
```

```python
#to check tha data types for all columns

df.dtypes
```

```python
#change data types of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

#let's confirm
df.dtypes
```

```python
#to check the value with highest gross - first line
df.sort_values(by =['gross'], inplace = False, ascending = False)
df 

#to check lowest gross - last line
df.tail()
```

```python
#drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
  df.drop_duplicates()
```

```python
#find possible correlation between some of the columns including year, score, votes, budget, gross and runtime

df[['year','score', 'votes', 'budget', 'gross', 'runtime']].corr() 

#uses pearson automatically, can specify other types using
#df.corr(methods='kendall') or df.corr(methods='spearman')
```

```python
#correllation value close to: +1 shows a large positive relationship, close to -1 shows a large negative relationship
#and close to 0 shows not relationship
#there's some positive relationship between gross and votes AND gross and budget
#to view the scatter plot of 'gross' and 'budget'

sns.regplot(x='budget', y ='gross', data =df, scatter_kws={"color" : "blue"}, line_kws={"color" : "red"})
plt.title('Budget vs gross earnings')
plt.ylim(0)
```

```python
#to view the correlation

correlation_matrix = df[['year','score', 'votes', 'budget', 'gross', 'runtime']].corr() 
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix')
plt.show()
```

```python
#to check for correlation between other columns that are not numeric, the columns can be converted in to the numeric form 
#and have unique idenfier so that they can be included in the correlation matrix

df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized

#every column not already numerized is now numerized
```
