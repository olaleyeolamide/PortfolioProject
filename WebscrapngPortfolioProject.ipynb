```python
from bs4 import BeautifulSoup
import requests 
```

```python
url = "https://www.gfmag.com/global-data/economic-data/largest-companies"

#get headers value from https://httpbin.org./get - specific for computer
#headers ={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36 Edg/117.0.2045.36"}

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
#print(soup)

soup1 = BeautifulSoup(soup.prettify(), 'html')
print(soup1)
#print(soup1)

#find the table

soup1.find_all('table')[1]
soup1.find('table', class_ = 'WORLD EVENTS VS STRONG FUNDAMENTALS')


```

```python
table = soup.find_all('table')[1]

print(table)
```

```python
tags = table.find_all('th')
print(tags)
```

```python
#extract all titles of each column from the table, strip to remove all whitespace and \n

maintitles = [title.text.strip() for title in tags]
print(maintitles)
```

```python
#create a dataframe

import pandas as pd

df = pd.DataFrame(columns = maintitles)
df
```

```python
col_data = table.find_all('tr')
```

```python
#invalid list value on the first and last row hence the split [1:11]
for row in col_data[1:11]:
    row_data = row.find_all('td')
    each_row = [data.text.strip() for data in row_data]
    print(each_row)
    
    length = len(df)
    df.loc[length] = each_row    
```

```python
df.drop_duplicates()
```

```python
#expost as csv, copy folder to export to, write filename.csv at the end. set index to False to remove the extra column containing the index number

df.to_csv(r"C:\Users\DELL\OneDrive\Desktop\Data Analysis\PortfolioWebscraping/company.csv", index=False)
```
